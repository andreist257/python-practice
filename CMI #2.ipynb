```python
#1. Write a function, add_it_up(), that takes a single integer as input and returns the sum of the integers from zero to the input parameter.
#The function should return 0 if a non-integer is passed in.

def add_it_up(integer):

  if type(integer) is int:
    L = list(range(0,integer + 1))
  else:
    L = [0]
  
  return sum(L)

print(add_it_up(10))

```

```python
#2. Cesar cipher. A Caesar cipher is a simple substitution cipher in which each letter of the plain text is substituted with a letter found by moving n places down the alphabet. 
#For example, assume the input plain text is the following: abcd xyz
#If the shift value, n, is 4, then the encrypted text would be the following: efgh bcd
#You are to write a function that accepts two arguments, a plain-text message and a number of letters to shift in the cipher. The function will return an encrypted string with all letters transformed and all punctuation and whitespace remaining unchanged.
#Note: You can assume the plain text is all lowercase ASCII except for whitespace and punctuation.

text = input('Insert your text here: ')

shift = int(input('Number of characters shifted: '))

def caesar_cipher(text, n_characters):
  
  shifted_text = ''

  for i in range(len(text)):

    if (ord(text[i]) < 65 or ord(text[i]) > 122) or (ord(text[i]) > 90 and ord(text[i]) < 97):  
      number = ord(text[i])
      letter = chr(number)
      shifted_text = shifted_text + letter
    elif (ord(text[i]) + n_characters >= 97 and ord(text[i]) + n_characters <= 122) or (ord(text[i]) + n_characters >= 65 and ord(text[i]) + n_characters <= 90):
      number = ord(text[i]) + n_characters
      letter = chr(number)
      shifted_text = shifted_text + letter
    elif (ord(text[i]) + n_characters < 97 and ord(text[i]) + n_characters > 90) or (ord(text[i]) + n_characters > 122):
      number = ord(text[i]) + n_characters - 26
      letter = chr(number)
      shifted_text = shifted_text + letter

  return shifted_text

print('Your shifted text is',caesar_cipher(text, shift))
```

```python
# Logparser. Accepts a filename on the command line. The file is a Linux-like log file from a system you are debugging. 
# Mixed in among the various statements are messages indicating the state of the device.
# The device state message has many possible values, but this program cares about only three: ON, OFF, and ERR.
# Your program will parse the given log file and print out a report giving how long the device was ON and the timestamp of any ERR conditions.

import pandas as pd
from datetime import datetime

data = pd.read_csv('logs.txt', delimiter = ' ', names = ['Month','Day','Time','Device Code', 'X', 'Y','Z','Status'])

data.drop(columns = ['Device Code','X','Y','Z'], inplace=True)

data['Timestamp'] = '2021' + ' ' + data['Month'] + ' ' + data['Day'].astype('str') + ' ' + data['Time']

data.drop(columns=['Month','Day','Time'], inplace=True)

data['Timestamp'] = data['Timestamp'].apply(lambda x: datetime.strptime(x, '%Y %b %d %H:%M:%S:%f'))
```

```python
time_diff_total = 0
for i in range(len(data)):
  if data.iloc[i].Status == 'ON':
    time_var = data.Timestamp.min()
    time_var1 = data.iloc[i].Timestamp
    time_diff_total = time_var1 - time_var
  else:
    continue

print('The device total working time was:',time_diff_total)
```

```python
error_logs = data[data['Status'] == 'ERR']

error_logs.to_csv('error_logs.txt')
```
